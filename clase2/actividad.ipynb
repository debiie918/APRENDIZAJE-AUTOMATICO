```python
import pandas as pd
#NOTA archivo 'inventario.xlsx' esta en la carpeta 'archivos semana 2'
dat_excel = pd.read_excel('archivos semana 2/inventario.xlsx')
dat_excel
```

3. b- Calcular el promedio de stock disponible para todos los productos.

```python
promedio_stock = dat_excel['stock'].mean()
print(f"El promedio de stock disponible es: {promedio_stock}")

# Se calcula el valor total del inventario (Precio * Stock de cada producto)
valor_total_inventario = (dat_excel['precio'] * dat_excel['stock']).sum()

# Calculamos la cantidad total de unidades en stock
unidades_totales = dat_excel['stock'].sum()

# Por ultimo el promedio del precio ponderado por el stock
promedio_precio_stock = valor_total_inventario / unidades_totales

print(f"Suma total de precios en stock: ${valor_total_inventario}")
print(f"Cantidad total de productos: {unidades_totales}")
print(f"El promedio del precio por producto en stock es: ${promedio_precio_stock:.2f}")
```

Combinacion de archivos y consultas

```python
# VENTAS- 
# Calculamos el total de unidades vendidas por cada producto
ventas_por_producto = dat_csv.groupby('producto')['cantidad'].sum().reset_index()

print(f"El total de unidades vendidas por producto es:\n{ventas_por_producto}")
```

ComparaciÃ³n entre ventas e inventario para calcular stock

```python
# Unimos especificando quÃ© columna usar de cada lado
ventas_inventario = pd.merge(
    ventas_por_producto, 
    dat_excel, 
    left_on='producto',   # El nombre en el archivo de la izquierda (ventas)
    right_on='nombre'     # El nombre en el archivo de la derecha (inventario)
)

# Calculo de cuÃ¡nto queda (Stock inicial - Cantidad vendida)
ventas_inventario['stock_final'] = ventas_inventario['stock'] - ventas_inventario['cantidad']
resultado = ventas_inventario[['nombre', 'stock', 'cantidad', 'stock_final']]
resultado.columns = ['Producto', 'Stock Inicial', 'Vendidos', 'Stock Restante']

print(resultado)
```

Â¿que producto no esta en el inventario?

```python
# El parÃ¡metro indicator=True nos dirÃ¡ si el dato estÃ¡ en ambas tablas o solo en la de ventas
verificacion = pd.merge(
    dat_csv, 
    dat_excel, 
    left_on='producto', 
    right_on='nombre', 
    how='left', 
    indicator=True
)

# Estos son los que NO se encontraron en el inventario
no_en_inventario = verificacion[verificacion['_merge'] == 'left_only']['producto'].unique()

# 3. Mostramos el resultado
if len(no_en_inventario) > 0:
    print("Los siguientes productos se vendieron pero NO estÃ¡n en el inventario:")
    for p in no_en_inventario:
        print(f"- {p}")
else:
    print("Todos los productos vendidos existen en el inventario.")
```
