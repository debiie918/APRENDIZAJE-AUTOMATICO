### Automobile.csv es un conjunto de datos que representa características de automóviles para la predicción de su consumo de combustible. Cada fila del conjunto de datos corresponde a un automóvil.

Tarea 1. Cargar el conjunto de datos en una estructura adecuada para su manipulación.

```python
import pandas as pd
dat_csv = pd.read_csv('archivo_clase3/Automobile.csv', encoding= 'ISO-8859-1')
dat_csv.head()
```

Tarea 2. Crear un gráfico de dispersión para visualizar las relaciones entre los pares de características numéricas `mpg` y `weight`

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot') 
fig, ax = plt.subplots(figsize=(10, 6)) 
ax.set_facecolor('#E0E0E0') # Codigo para un fondo gris claro.
ax.grid(False) # Desactiva las líneas de la cuadrícula.
plt.scatter(dat_csv['weight'], dat_csv['mpg'], color='pink', edgecolors='deeppink', alpha=0.7) 
plt.title('Relación entre Peso (Weight) y Rendimiento (MPG)', fontsize=14, fontweight='bold') 
plt.xlabel('Peso del Vehículo (Libras)', fontsize=12) 
plt.ylabel('Millas por Galón (MPG)', fontsize=12)

plt.xlabel('Peso (Weight)')
plt.ylabel('Milla por galeon (MPG)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Configuramos un estilo elegante
sns.set_theme(style="dark") 
# Creamos el gráfico indicando las columnas y el DataFrame
sns.scatterplot(data=dat_csv, x='weight', y='mpg', color='deeppink')
plt.title('Relación entre Peso (Weight) y Rendimiento (MPG)', fontsize=14, fontweight='bold')
plt.show()
```

Tarea 3. Generar un histograma del consumo de combustible (`mpg`) para analizar su distribución.

```python

import matplotlib.pyplot as plt
import seaborn as sns

# 2. Configurar la figura y el fondo gris que pediste
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor('#E0E0E0')  # Fondo gris
ax.grid(False)                # Sin cuadrícula

sns.histplot(dat_csv['mpg'], bins=30, kde=True, color='deeppink', ax=ax) #crear el histograma usando Seaborn

plt.title('Distribución del Consumo de Combustible (MPG)', fontsize=14, fontweight='bold')
plt.xlabel('Millas por Galón (MPG)', fontsize=12)
plt.ylabel('Frecuencia (Cantidad de autos)', fontsize=12)

# 5. Mostrar el gráfico
plt.show()
```

Tarea 4. Graficar un diagrama de caja (boxplot) de las características `cylinders` y `origin` para identificar posibles valores atípicos.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# aplico la misma configuración de figura y fondo que anteriores graficos
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor('#E0E0E0')  # Fondo gris liso
ax.grid(False)                # Sin cuadrícula

sns.boxplot(data=dat_csv, x='origin', y='cylinders', palette='Pastel1', ax=ax) #boxplot

plt.title('Distribución de Cilindros por Origen del Vehículo', fontsize=14, fontweight='bold')
plt.xlabel('Origen', fontsize=12)
plt.ylabel('Numero de Cilindros', fontsize=12)
plt.show()
```

Tarea 5. Realizar un gráfico de barras que muestre la distribución de automóviles según su año de modelo (`model_year`) y origen (`origin`).

```python
import matplotlib.pyplot as plt
import seaborn as sns

# figura y el estilo
fig, ax = plt.subplots(figsize=(12, 6))
ax.set_facecolor('#E0E0E0')  # Fondo gris
ax.grid(False)                # Sin cuadrícula

sns.countplot(data=dat_csv, x='model_year', hue='origin', palette='magma', ax=ax)

plt.title('Distribución de Automóviles por Año', fontsize=14, fontweight='bold')
plt.xlabel('Año del Modelo', fontsize=12)
plt.ylabel('Cantidad de Vehículos', fontsize=12)

plt.show()
```
